```python
import re
import requests
from bs4 import BeautifulSoup

news_title=[]

for n in range(1,100):
    req = requests.get('https://search.naver.com/search.naver?query=%EC%A0%84%EB%8F%99+%ED%82%A5%EB%B3%B4%EB%93%9C&where=news&ie=utf8&sm=nws_hty'+str(n))
    html = req.content
    html = html.decode('utf-8')
    bs = BeautifulSoup(html, 'html.parser')

    title = bs.select('div > div > a')

    for titledom in title:
        news_title.append(titledom.text.strip())


print(news_title)
```
